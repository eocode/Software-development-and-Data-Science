````python
columns = data.readline().strip().split(',')
columns
````

````python
n_columns = len(columns)
n_columns
````

````python
dataframe_dict = {element:[] for element in columns}
dataframe_dict
````

````python
counter = 0
for line in data: 
    values = line.strip().split(',')
    for i in range(len(columns)):
        dataframe_dict[columns[i]].append(values[i])
    counter += 1

print("El dataset tiene", counter, "filas y", n_columns, "columnas")
````
