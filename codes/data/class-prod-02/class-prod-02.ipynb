# Rellenar valores nulos

Los valores pueden faltar por errores en la recolección de datos y en la extracción de los datos

```python
pd.isnull(data["age"])
```

```python
pd.isnull(data["age"]).values
```

```python
pd.isnull(data["age"]).values.sum()
```

```python
data.shape
```

```python
data.dropna(axis=0, how="all") # Borra todas las filas con valores nulos en todas las columnas
```

```python
data.dropna(axis=0, how="any") # Borra todas las filas con valores nulos en alguna de las columnas
```

```python
data.dropna(subset=["age"])
```

```python
data.fillna(0) # Rellena los valores nulos con un 0
```

```python
data["age"].fillna(data["age"].mean()) # Rellena los valores nulos en una columna por la media de dicha columna
```

```python
data["age"].fillna(method="ffill") #Reemplaza los valores nulos por el valor más cercano hacia atrás
# ffill = Forward fill
```

```python
data["age"].fillna(method="backfill") #Reemplaza los valores nulos por el valor más cercano hacia adelante
```

```python
data = data.dropna(subset=["age"])
```

```python
%matplotlib inline
```

## Histograma de frecuencias

El histograma de frecuencias nos permite ver qué valores de una determinada columna aparecen más que otros

```python
plt.hist(data["age"]) # bins es la cantidad de divisiones
# plt.hist(data["Day Calls"], bins = [0,60,90,120,150,180]) # Los datos que están en los rangos se unen en una sola barra
# k = int(np.ceil(1+np.log2(3333))) # Regla de Sturges
# plt.hist(data["age"], bins = k)
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Histograma de frecuencia de las edades")
```

## Boxplot

El boxplot, o diagrama de "caja y bigotes", nos permite ver de forma más detallada cuáles son los valores que están más cerca de la mediana y los que están más alejados de la mediana

```python
plt.boxplot(data["age"])
plt.ylabel("Edades")
plt.title("Boxplot de las edades")

# Las pelotitas son llamadas outliers (valores atípicos)
```

```python
data["age"].describe()
```

```python
# Rango intercuartílico = Altura de la caja del boxplot

#IQR = 39 -21
IQR = data["age"].quantile(0.75) - data["age"].quantile(0.25)
IQR
```

```python
# El bigote inferior llega hasta...

data["age"].quantile(0.25) - 1.5*IQR
```

```python
# El bigote superior llega hasta...

data["age"].quantile(0.75) + 1.5*IQR
```
